```python
def bordes_zonas_oscuras(imagen_path, umbral_oscuridad=60, canny_umbral1=120, canny_umbral2=250, mostrar=False):
    """
    Extrae los bordes de las regiones oscuras de una imagen.

    - umbral_oscuridad: pÃ­xeles con valor menor a esto se consideran "oscuros".
    """
    img = cv2.imread(imagen_path, cv2.IMREAD_GRAYSCALE)
    img_rgb = cv2.imread(imagen_path, cv2.IMREAD_COLOR_RGB)
    # img_rgb = Image.fromarray(cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB))  # Convert to PIL image
    # img_rgb = img_rgb.point(lambda p: p > 128 and 255)
    # img_rgb = cv2.cvtColor(np.array(img_rgb), cv2.COLOR_BGR2RGB)
    
    if img is None:
        raise FileNotFoundError(f"No se pudo cargar la imagen desde {imagen_path}")
    
    # Crear una mÃ¡scara binaria de regiones oscuras
    mascara_oscura = cv2.threshold(img, umbral_oscuridad, 255, cv2.THRESH_BINARY_INV)[1]
    
    # Aplicar Canny solo a las regiones oscuras
    regiones_oscuras = cv2.bitwise_and(img, img, mask=mascara_oscura)
    bordes = cv2.Canny(regiones_oscuras, canny_umbral1, canny_umbral2)

    if mostrar:
        plt.figure(figsize=(12, 4))

        plt.subplot(1, 3, 1)
        plt.title('Original')
        #plt.imshow(img, cmap='gray')
        plt.imshow(img_rgb)
        plt.axis('off')

        plt.subplot(1, 3, 2)
        plt.title(f'MÃ¡scara Oscura < {umbral_oscuridad}')
        #plt.imshow(mascara_oscura, cmap='gray')
        plt.imshow(img, cmap='gray')
        plt.axis('off')

        plt.subplot(1, 3, 3)
        plt.title('Bordes en Zonas Oscuras')
        plt.imshow(bordes, cmap='gray')
        plt.axis('off')

        plt.tight_layout()
        plt.show()

    return bordes


bordes = Image.fromarray(bordes_zonas_oscuras(img_path))
bordes

```

```python
def zonas_dentro_de_bordes_oscuros(imagen_path, umbral_oscuridad=60, canny_umbral1=120, canny_umbral2=250, mostrar=False):
    """
    Extrae las zonas dentro de los bordes de regiones oscuras de una imagen.
    """
    img = cv2.imread(imagen_path, cv2.IMREAD_GRAYSCALE)
    img_color = cv2.cvtColor(cv2.imread(imagen_path), cv2.COLOR_BGR2RGB)  # Imagen en color en formato RGB

    if img is None:
        raise FileNotFoundError(f"No se pudo cargar la imagen desde {imagen_path}")

    # Crear una mÃ¡scara binaria de regiones oscuras
    mascara_oscura = cv2.threshold(img, umbral_oscuridad, 255, cv2.THRESH_BINARY_INV)[1]

    # Aplicar Canny solo a regiones oscuras
    regiones_oscuras = cv2.bitwise_and(img, img, mask=mascara_oscura)
    bordes = cv2.Canny(regiones_oscuras, canny_umbral1, canny_umbral2)

    # Cerrar los bordes para formar regiones completas
    kernel = np.ones((5, 5), np.uint8)
    bordes_cerrados = cv2.morphologyEx(bordes, cv2.MORPH_CLOSE, kernel)

    # Encontrar contornos
    contornos, _ = cv2.findContours(bordes_cerrados, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Crear una nueva mÃ¡scara y rellenar los contornos
    mascara_rellena = np.zeros_like(img)
    cv2.drawContours(mascara_rellena, contornos, -1, color=255, thickness=cv2.FILLED)

    # Aplicar la mÃ¡scara rellena a la imagen original en color
    resultado = cv2.bitwise_and(img_color, img_color, mask=mascara_rellena)

    if mostrar:
        plt.figure(figsize=(12, 4))

        plt.subplot(1, 4, 1)
        plt.title('Original RGB')
        plt.imshow(img_color)
        plt.axis('off')

        plt.subplot(1, 4, 2)
        plt.title('Regiones Oscuras')
        plt.imshow(mascara_oscura, cmap='gray')
        plt.axis('off')

        plt.subplot(1, 4, 3)
        plt.title('Bordes Cerrados')
        plt.imshow(bordes_cerrados, cmap='gray')
        plt.axis('off')

        plt.subplot(1, 4, 4)
        plt.title('Zonas Dentro de Bordes')
        plt.imshow(resultado)
        plt.axis('off')

        plt.tight_layout()
        plt.show()

    return resultado
Image.fromarray(zonas_dentro_de_bordes_oscuros(img_path, mostrar=True))
```

```python
def recortar_centrado(imagen, nuevo_tamano):
    """
    Recorta la imagen para que quede centrada con el tamaÃ±o especificado.

    Args:
    - imagen: PIL.Image.Image, la imagen a recortar.
    - nuevo_tamano: tuple, el tamaÃ±o deseado (ancho, alto).

    Returns:
    - PIL.Image.Image, la imagen recortada.
    """
    ancho, alto = imagen.size
    nuevo_ancho, nuevo_alto = nuevo_tamano

    # Calcular las coordenadas para el recorte centrado
    izquierda = (ancho - nuevo_ancho) // 2
    superior = (alto - nuevo_alto) // 2
    derecha = izquierda + nuevo_ancho
    inferior = superior + nuevo_alto

    return imagen.crop((izquierda, superior, derecha, inferior))

# Recortar la imagen bordes al tamaÃ±o 224x224
bordes_centrado = recortar_centrado(bordes, (224*2, 224*2))
bordes_centrado
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def extraer_bordes(imagen_path, umbral1=100, umbral2=200, mostrar=False):
    """
    Extrae los bordes de una imagen (espectrograma) usando el detector de Canny.

    ParÃ¡metros:
    - imagen_path: ruta de la imagen del espectrograma.
    - umbral1: primer umbral para el algoritmo de Canny.
    - umbral2: segundo umbral para el algoritmo de Canny.
    - mostrar: si es True, muestra la imagen original y los bordes.

    Retorna:
    - edges: imagen binaria con los bordes detectados.
    """
    # Cargar la imagen en escala de grises
    img = cv2.imread(imagen_path, cv2.IMREAD_COLOR_RGB)
    
    if img is None:
        raise FileNotFoundError(f"No se pudo cargar la imagen desde {imagen_path}")
    
    # Aplicar el detector de bordes de Canny
    edges = cv2.Canny(img, umbral1, umbral2)
    
    if mostrar:
        plt.figure(figsize=(10, 5))
        plt.subplot(1, 2, 1)
        plt.title('Original')
        plt.imshow(img, cmap='gray')
        plt.axis('off')

        plt.subplot(1, 2, 2)
        plt.title('Bordes con Canny')
        plt.imshow(edges, cmap='gray')
        plt.axis('off')
        plt.tight_layout()
        plt.show()

    return edges
edges = extraer_bordes(df.iloc[idx]['path'], umbral1=255, umbral2=500, mostrar=True)
```
