```python
total_iter= len(pkl_files)
progress_bar = tqdm(total=total_iter, desc="Progress", unit="iter")
df = pd.DataFrame()
with open('event_labels/performance_labels', 'rb') as file:
     labels_array = pickle.load(file)

reaction_times = []
for df_temp in labels_array:
    reaction_times.extend(df_temp['reaction times'].values)

#reaction_times = np.array(reaction_times)
mean_reaction_time = np.mean(reaction_times)
percentiles = np.percentile(reaction_times, [45, 50, 55])
percentiles_2 = np.percentile(reaction_times, [40, 50, 60])

percentile_45 = percentiles[0]
percentile_55 = percentiles[2]
percentile_45_2 = percentiles_2[0]
percentile_55_2 = percentiles_2[2]
    
for i, pkl in enumerate(pkl_files):
    labels_df = labels_array[i]

    with open(pkl, 'rb') as file:
        events = pickle.load(file)
    filtered_events = [item for i, item in enumerate(events[1:]) if i in labels_df.index]
    
    labels_df['drop 10'] = labels_df['reaction times'].apply(lambda x: 0 if x < percentile_45 else (1 if x > percentile_55 else np.nan))
    labels_df['drop 20'] = labels_df['reaction times'].apply(lambda x: 0 if x < percentile_45_2 else (1 if x > percentile_55_2 else np.nan))
    
    df_temp = pd.DataFrame()
    for j, event in enumerate(filtered_events):
        label = labels_df['label'].values[j]
        drop_10 = labels_df['drop 10'].values[j]
        drop_20 = labels_df['drop 20'].values[j]
        event_idx = labels_df.index.values[j]
        data = pd.DataFrame(event.get_data())
        data = data.transpose()
        data['class'] = int(pkl[-5])
        data['subject'] = int(pkl[8:10])
        data['label'] = label
        data['drop 10'] = drop_10
        data['drop 20'] = drop_20
        data['event'] = event_idx
        df_temp = pd.concat([df_temp, data], ignore_index=True)
        #print(' class: {}, subject: {}, label: {}'.format(int(pkl[-5]), int(pkl[8:10]), label))

    df = pd.concat([df, df_temp], ignore_index=True)
    progress_bar.update(1)

# Cerrar la barra de progreso al finalizar
progress_bar.close()

df

```
